# Feature Engineering Process
Step-by-step feature creation for moving averages, volatility, and lag features.

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load Raw Data
df = pd.read_csv('../data/raw_data.csv', parse_dates=['Date'], index_col='Date')
df.head()
```

## 1. Handling Missing Values

```
df.ffill(inplace=True)
df.bfill(inplace=True)
df.isnull().sum()
```

## 2. Advanced Feature Generation (MA, Return, Volatility)

```
df['MA_7'] = df['Close'].rolling(window=7).mean()
df['MA_30'] = df['Close'].rolling(window=30).mean()
df['Daily_Return'] = df['Close'].pct_change()
df['Rolling_Volatility_30'] = df['Daily_Return'].rolling(window=30).std()
df['Lag_1'] = df['Close'].shift(1)
df.dropna(inplace=True)
df.head()
```

## 3. Data Normalization

```
scaler = MinMaxScaler()
# Attempting to gather relevant numeric columns for scaling
cols = ['Close', 'MA_7', 'Rolling_Volatility_30']
scaled_data = scaler.fit_transform(df[cols])
for i, c in enumerate(cols):
    df[f'Norm_{c}'] = scaled_data[:, i]
df.head()
```
